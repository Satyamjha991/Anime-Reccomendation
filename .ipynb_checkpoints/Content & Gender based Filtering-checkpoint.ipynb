```python
def build_chart_on_members(Gender, percentile=0.85):
    df = gen_anime_data[gen_anime_data['Genders'] == Gender]
    V_dat=df["Members"]
    C=df["Score"].mean()
    R_dat=df["Score"]
    M=V_dat.quantile(percentile)
    
    qualified = df[(df['Members'] >= M)][["anime_id", "Name","Score","Genders","Popularity","Members","Favorites","Plan to Watch","Duration","Episodes","Studios"]]
                   
                   
    #qualified=df[["anime_id", "Name","Score","Genders","Popularity","Members","Favorites","Plan to Watch","Duration","Episodes","Studios"]].sort_values(by='Score', ascending=False).query('Members>=M')
    qualified['wr'] = ((V_dat/(V_dat+M))*R_dat) +((M/(M+V_dat))*C)
    qualified.sort_values('wr', ascending=False).head(250)
    return qualified
```

```python
def build_chart_on_favourites(Gender, percentile=0.85):
    df = gen_anime_data[gen_anime_data['Genders'] == Gender]
    V_dat=df["Favorites"]
    C=df["Score"].mean()
    R_dat=df["Score"]
    M=V_dat.quantile(percentile)
    
    qualified = df[(df['Favorites'] >= M)][["anime_id", "Name","Score","Genders","Popularity","Members","Favorites","Plan to Watch","Duration","Episodes","Studios"]]
    qualified['wr'] = ((V_dat/(V_dat+M))*R_dat) +((M/(M+V_dat))*C)
    qualified.sort_values('wr', ascending=False).head(250)
    return qualified
```

```python
def build_chart_on_votes(Gender, percentile=0.85):
    df = gen_anime_data[gen_anime_data['Genders'] == Gender]
    V_dat=df["total_rating_count"]
    C=df["Score"].mean()
    R_dat=df["Score"]
    M=V_dat.quantile(percentile)
    
    qualified = df[(df['total_rating_count'] >= M) & (df['total_rating_count'].notnull()) ][["anime_id", "Name","Score","Genders","Popularity","Members","Favorites","Plan to Watch","Duration","Episodes","Studios"]]
    qualified['wr'] = ((V_dat/(V_dat+M))*R_dat) +((M/(M+V_dat))*C)
    qualified.sort_values('wr', ascending=False).head(250)
    return qualified
```

```python
import re
def text_cleaning(text):
    text = re.sub(r'&quot;', '', text)
    text = re.sub(r'.hack//', '', text)
    text = re.sub(r'&#039;', '', text)
    text = re.sub(r'A&#039;s', '', text)
    text = re.sub(r'I&#039;', 'I\'', text)
    text = re.sub(r'&amp;', 'and', text)
    return text
```

```python
anime_content["Name"]=anime_content["Name"].apply(text_cleaning)
anime_content["synopsis"]=anime_content["synopsis"].apply(text_cleaning)
anime_content["Genders"]=anime_content["Genders"].apply(text_cleaning)
```

```python
def improved_recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:30]
    anime_indices = [i[0] for i in sim_scores]
    df = anime_content.iloc[anime_indices][['Name', 'Score', 'Genders', 'total_rating_count','English name']]
    V_dat=df["total_rating_count"]
    C=df["Score"].mean()
    R_dat=df["Score"]
    M=V_dat.quantile(0.5)
    qualified = df[(df['total_rating_count'] >= M) & (df['total_rating_count'].notnull()) ][[ "Name","Score","Genders","English name"]]
    qualified['wr'] = ((V_dat/(V_dat+M))*R_dat) +((M/(M+V_dat))*C)
    return qualified.sort_values('wr', ascending=False).head(10)
```
